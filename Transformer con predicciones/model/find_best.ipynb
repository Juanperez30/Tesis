```python
import numpy as np
import subprocess
```

```python
learning_rate = [0.2,0.5,0.8,0.02,0.002]
lr_decay = [0.2,0.08,0.002,0.5]
epochs = [400,600]
batches = [10,50,100]
length_attention = [6,10,30]

parameters_a = []
RMSE_a = []
```

```python
r = []
for lr in learning_rate:
    for lrdecay in lr_decay:
        for epoch in epochs:
            for batch in batches:
                for attention in length_attention:
                                RMSE = 0                             
                                parameters_a.append([lr, lrdecay, epoch, batch, attention])
                                command = f"python driver.py --dataset ../basic_series/raw_12.csv --lr {lr} --lr_decay {lrdecay} --epochs {epoch} --dy 15 --length {attention} --batch {batch}"                    
                                try:         
                                    print("######### Entered training ##########", lr ,' ', lrdecay, ' ', epoch, ' ', batch, ' ', attention)                                                                       
                                    completed_process = subprocess.run(command, shell=True, check=True, stdout=subprocess.PIPE,stderr=subprocess.PIPE)
                                    
                                    RMSE = float(completed_process.stdout.decode().strip())
                                    print(f"RMSE: {RMSE}")
                                                                        
                                    print("######### Finished training ##########", lr , lrdecay, epoch)
                                except Exception as e:                            
                                    print(f"Error running command: {command}") 
                                    print(e)    
                                    RMSE_a.append(RMSE)                                                       
                                    continue  
                                RMSE_a.append(RMSE)                        
```

```python
np.save('parameters_list.npy', parameters_a)
np.save('RMSE_list.npy',RMSE_a)
```

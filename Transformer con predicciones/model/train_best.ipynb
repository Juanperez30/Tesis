```python
def clean_dict_values(input_dict):
    cleaned_dict = {}

    for key, value in input_dict.items():
        if isinstance(value, str):
            # Remove newline characters and split the string into a list of floats
            value = value.replace('\r', '').replace('\n', '')
            value = [float(val) for val in value.strip('[]').split()]
        cleaned_dict[key] = value

    return cleaned_dict


cleaned_dict = clean_dict_values(predicciones)

print(cleaned_dict)
```

```python
def find_max_and_round(dict_data):
    max_values = {}

    for key, values in dict_data.items():
        if isinstance(values, list):
            max_value = max(values)
            if not max_value.is_integer():
                max_values[key] = round(max_value) + 1  # Round to the next integer
            else:
                max_values[key] = int(max_value)
        else:
            max_values[key] = values  # Keep non-list values as they are

    return max_values

max_rounded_dict = find_max_and_round(cleaned_dict)

print(max_rounded_dict)
```
